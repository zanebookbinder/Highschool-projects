```python
from PIL import Image
import numpy as np
import os, os.path, time
import pandas as pd
import math
```

```python
df = pd.read_csv('??/annotation.txt')
df.head()
```

```python
count = 0
for path in df['Path']:
    df['Path'][count] = path.replace('./characters/','the-simpsons-characters-dataset/simpsons_dataset/',1)
    count+=1
df.head()
```

```python
print(df['Path'][1100])
img_file = Image.open(df['Path'][1])
display(img_file)
```

```python
rowslist = []
imagenames = []
pathnames = []

count = 0
for file in df['Path']:
    img_file = Image.open(file) #open image

    width, height = img_file.size
    if width/height < 1.5 and width/height > .7:
        img_file = img_file.resize((100,100), Image.ANTIALIAS)
        #display(img_file)
        img_grey = img_file.convert('L') #greyscale
        #display(img_grey) #display images if needed
        value = np.asarray(img_grey.getdata(),dtype=np.int) #numpy array
        print(type(img_grey.getdata()))
        rowslist.append(value)
        imagenames.append(df['Person'][count])
        pathnames.append(df['Path'][count])
    
    count +=1
```

```python
df = pd.DataFrame(rowslist)               
```

```python
df = df/255
df.head()
```

```python
df['Person'] = imagenames
df['Path'] = pathnames
df.head(3)
```

```python
df.to_csv('all_images.csv',index=False)
```
