```python
import pandas
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from keras.utils import np_utils
import math
import numpy as np
```

```python
df = pandas.read_csv("NBA_data.csv")
df = df.drop(['Hand (Length)','Hand (Width)','Player','Year'], axis=1)
df.head()
```

```python
picks = np.empty([0,0])
df['Draft pick'] = df['Draft pick'].fillna(0)
for player in np.nditer(df['Draft pick']):
    if player > 3 or player == 0:
        player = 0
    else:
        player = 1
    picks = np.append(picks,[player])
df['Draft pick'] = picks
df = df.fillna(0)
df.head()
```

```python
Y=df['Draft pick']
X=df.loc[:, df.columns != 'Draft pick']
```

```python
x = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X = pandas.DataFrame(x_scaled)
```

```python
def create_baseline():
    # create model
    model = Sequential()
    model.add(Dense(25, input_dim=13, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(12, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(7, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=create_baseline, epochs=300, batch_size=6, verbose=0)
kfold = StratifiedKFold(n_splits=3, shuffle=True)
results = cross_val_score(estimator, X, Y,cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```python
model = create_baseline()
model.fit(X,Y,epochs=300,batch_size=6,verbose=0)
```

```python
for num in range(300):
    test = X.iloc[num]
    test = pandas.DataFrame(test)
    prediction = model.predict(test.T)
    #if prediction[0][0] > 0.05:
        #print('Your predicted draft is {}'.format(prediction[0][0]))
```
