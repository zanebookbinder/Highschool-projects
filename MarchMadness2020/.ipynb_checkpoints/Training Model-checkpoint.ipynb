```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense,Dropout,Activation
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from keras.utils import np_utils
import math
import numpy as np
import random
```

```python
averages = pd.read_csv('teamaverages.csv')
averages.head()
```

```python
margins = []
for index, row in averages.iterrows():
    margin = row['TScore'] - row['AScore']
    margins.append(margin)
averages['Margin'] = margins
averages.head()
```

```python
cresults = pd.read_csv('data/MDataFiles_Stage1/MNCAATourneyCompactResults.csv').drop(['NumOT','DayNum'],axis=1)
cresults = cresults[cresults['Season'] >= 2003].reset_index(drop=True)
cresults.head()
```

```python
averages.columns
```

```python
cols1 = averages.columns.tolist()
cols2 = ["2" + stat for stat in averages.columns.tolist()]
cols = cols1 + cols2
gamemargins = []

addstats = pd.DataFrame(columns = cols)
for index,row in cresults.iterrows():
    var = random.choice([True, False])
    year = row['Season']
    wid = row['WTeamID']
    lid = row['LTeamID']
    wstats = averages[averages['Season'] == year]
    wstats = wstats[wstats['TeamID'] == wid].values.tolist()[0]
    lstats = averages[averages['Season'] == year]
    lstats = lstats[lstats['TeamID'] == lid].values.tolist()[0]
    if var == True:
        stats = wstats + lstats
        gmargin = row['WScore'] - row['LScore']
    else:
        stats = lstats + wstats
        gmargin = row['LScore'] - row['WScore']
    gamemargins.append(gmargin)
    a_len = len(addstats)
    addstats.loc[a_len] = stats
cresults['TourneyMargin'] = gamemargins
```

```python
addstats = addstats.drop(['Season','2Season','TeamID','2TeamID'],axis=1)
addstats.columns
```

```python
cresults = cresults.reset_index(drop=True).drop(['WLoc','WScore','LScore'],axis=1)
alldata = pd.concat([cresults, addstats], axis=1)
alldata.head()
```

```python
for index, row in alldata.iterrows():
    if row['TourneyMargin'] > 0:
        alldata.at[index,'TourneyMargin'] = 1
    if row['TourneyMargin'] < 0:
        alldata.at[index,'TourneyMargin'] = 0
alldata.head(2)

usefuldata = alldata[['Season', 'TScore', 'AScore',
       'TFGM', 'TFGA', 'TFGM3', 'TFGA3', 'TFTM', 'TFTA', 'TOR', 'TDR', 'TAst',
       'TTO', 'TStl', 'TBlk', 'TPF', 'AFGM', 'AFGA', 'AFGM3', 'AFGA3', 'AFTM',
       'AFTA', 'AOR', 'ADR', 'AAst', 'ATO', 'AStl', 'ABlk', 'APF', 'Margin',
       '2TScore', '2AScore', '2TFGM', '2TFGA', '2TFGM3', '2TFGA3',
       '2TFTM', '2TFTA', '2TOR', '2TDR', '2TAst', '2TTO', '2TStl', '2TBlk',
       '2TPF', '2AFGM', '2AFGA', '2AFGM3', '2AFGA3', '2AFTM', '2AFTA', '2AOR',
       '2ADR', '2AAst', '2ATO', '2AStl', '2ABlk', '2APF', '2Margin','TourneyMargin']]
```

```python
train = usefuldata[usefuldata['Season'] < 2015].drop(['Season'],axis=1)
test = usefuldata[usefuldata['Season'] < 2015].drop(['Season'],axis=1)
X_train = train.loc[:, train.columns != 'TourneyMargin']
Y_train = train['TourneyMargin']

X_test = test.loc[:, test.columns != 'TourneyMargin']
Y_test = test['TourneyMargin']
```

```python
x = X_train.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_train = pd.DataFrame(x_scaled)

x = X_test.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X_test = pd.DataFrame(x_scaled)
```

```python
def create_baseline():
    model = Sequential()
    model.add(Dense(100, input_dim=58, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(60, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(30, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(15, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(7, activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = create_baseline()
model.fit(X_train, Y_train, batch_size = 5, epochs = 1000, verbose = 2)
```

```python
model.save('finalmodel.h5')
```
