```python
from keras.models import load_model
import pandas as pd
```

```python
model = load_model('finalmodel.h5')
```

```python
ateams = pd.read_csv('data/MDataFiles_Stage1/MNCAATourneySeeds.csv').drop(['Seed'],axis=1)
ateams = ateams[ateams['Season'] > 2014].reset_index(drop=True)
ateams.head()
```

```python
averages = pd.read_csv('teamaverages.csv')
margins = []
for index, row in averages.iterrows():
    margin = row['TScore'] - row['AScore']
    margins.append(margin)
averages['Margin'] = margins
```

```python
averages.columns
```

```python
cols = averages.columns

fifteen = pd.DataFrame(columns = cols)
sixteen = pd.DataFrame(columns = cols)
seventeen = pd.DataFrame(columns = cols)
eighteen = pd.DataFrame(columns = cols)
nineteen = pd.DataFrame(columns = cols)
years = [fifteen, sixteen, seventeen, eighteen, nineteen]

for index, row in ateams.iterrows():
    year = row['Season']
    ID = row['TeamID']
    sstats = list(averages[(averages['Season'] == year) & (averages['TeamID'] == ID)].reset_index(drop = True).loc[0])
    if year == 2015:
        fifteen.loc[len(fifteen)] = sstats
    if year == 2016:
        sixteen.loc[len(sixteen)] = sstats
    if year == 2017:
        seventeen.loc[len(seventeen)] = sstats
    if year == 2018:
        eighteen.loc[len(eighteen)] = sstats
    if year == 2019:
        nineteen.loc[len(nineteen)] = sstats
```

```python
nineteen.head()
```

```python
games = pd.read_csv('data/MDataFiles_Stage1/MNCAATourneyCompactResults.csv').drop(['NumOT','DayNum','WLoc'],axis=1)
games = games[games['Season'] > 2014].reset_index(drop=True)
games.head(2)
```

```python
def makeloserfirst(df):
    output = pd.DataFrame(columns = ['Season','TeamID1','1Score','TeamID2','2Score'])
    for index, row in df.iterrows():
        if row['WTeamID'] > row['LTeamID']:
            row = list(row[['Season','LTeamID','LScore','WTeamID','WScore']])
        else:
            row = list(row[['Season','WTeamID','WScore','LTeamID','LScore']])
        output.loc[len(output)] = row
    return output
ordered = makeloserfirst(games)
```

```python
finalcols = ['TeamID', 'TScore', 'AScore', 'TFGM', 'TFGA',
       'TFGM3', 'TFGA3', 'TFTM', 'TFTA', 'TOR', 'TDR', 'TAst', 'TTO', 'TStl',
       'TBlk', 'TPF', 'AFGM', 'AFGA', 'AFGM3', 'AFGA3', 'AFTM', 'AFTA', 'AOR',
       'ADR', 'AAst', 'ATO', 'AStl', 'ABlk', 'APF', 'Margin']

cols2 = finalcols + ["2" + stat for stat in finalcols]
revcols2 = ["2" + stat for stat in finalcols] + finalcols
```

I HAVE TO CONNECT GAMES WITH THE YEARLY DATASETS(FIFTEEN, SIXTEEN, ETC.) SO THAT I AM ADDING THE TEAM STATS TO FINALDATA

```python
dictionary = {2015:fifteen,2016:sixteen,2017:seventeen,2018:eighteen,2019:nineteen}

finaldata = pd.DataFrame(columns = cols2)
for index, row in ordered.iterrows():
    yeardf = dictionary[row['Season']]
    av1 = yeardf[yeardf['TeamID'] == row['TeamID1']].drop(['Season'],axis=1).values.tolist()[0]
    av2 = yeardf[yeardf['TeamID'] == row['TeamID2']].drop(['Season'],axis=1).values.tolist()[0]
    av = av1+av2
    finaldata.loc[len(finaldata)] = av
```

```python
ids = finaldata[['TeamID','2TeamID']]
finaldata = finaldata.drop(['TeamID','2TeamID'],axis=1)
finaldata.head(5)
```

```python
results = model.predict(finaldata)
```

```python
finalresults = []
for num in results:
    num = round(num[0],7)
    finalresults.append(num)
```
